```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Jupyter Notebook Magic Command for crisp charts
%matplotlib inline

# Configuration & Styling
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'font.size': 11,
    'axes.labelsize': 11,
    'axes.titlesize': 13,
    'figure.titlesize': 16
})

# Custom Corporate Palette
custom_palette = ['#2b5c8f', '#d95f02']

# Load Dataset
file_name = "/content/WA_Fn-UseC_-HR-Employee-Attrition.csv"
if not os.path.exists(file_name):
    raise FileNotFoundError(f"Could not find '{file_name}'. Please ensure it is in the same folder as this notebook.")

df = pd.read_csv(file_name)
```

```python
# Demographics & Diversity Insights
fig1, axes1 = plt.subplots(1, 3, figsize=(18, 5))
fig1.suptitle("Dashboard 1: Demographics & Diversity Insights", y=1.02, fontweight='bold', color='#1a1a1a')

sns.histplot(data=df, x='Age', hue='Attrition', multiple='stack', ax=axes1[0], palette=custom_palette, kde=True)
axes1[0].set_title('Age Distribution vs Attrition')

sns.countplot(data=df, x='Department', hue='Gender', ax=axes1[1], palette='Set2')
axes1[1].set_title('Gender Distribution by Department')

sns.countplot(data=df, x='MaritalStatus', hue='Attrition', ax=axes1[2], palette=custom_palette)
axes1[2].set_title('Marital Status vs Attrition')

plt.tight_layout()
plt.show()
```

```python
# Compensation & Financial Drivers
fig2, axes2 = plt.subplots(1, 3, figsize=(18, 5))
fig2.suptitle("Dashboard 2: Compensation & Financial Drivers", y=1.02, fontweight='bold', color='#1a1a1a')

sns.scatterplot(data=df, x='Age', y='MonthlyIncome', hue='Attrition', alpha=0.7, ax=axes2[0], palette=custom_palette)
axes2[0].set_title('Income vs Age Trend')
axes2[0].set_ylabel('Monthly Income ($)')

sns.boxplot(data=df, x='Attrition', y='PercentSalaryHike', ax=axes2[1], palette=custom_palette)
axes2[1].set_title('Salary Hike % vs Attrition')

sns.countplot(data=df, x='StockOptionLevel', hue='Attrition', ax=axes2[2], palette=custom_palette)
axes2[2].set_title('Stock Option Level vs Attrition')

plt.tight_layout()
plt.show()
```

```python
# Work-Life Balance & Burnout Factors
fig3, axes3 = plt.subplots(1, 3, figsize=(18, 5))
fig3.suptitle("Dashboard 3: Work-Life Balance & Burnout Factors", y=1.02, fontweight='bold', color='#1a1a1a')

sns.countplot(data=df, x='BusinessTravel', hue='Attrition', ax=axes3[0], palette=custom_palette)
axes3[0].set_title('Business Travel Frequency')

sns.boxplot(data=df, x='Attrition', y='DistanceFromHome', ax=axes3[1], palette=custom_palette)
axes3[1].set_title('Commute Distance (Miles)')
axes3[1].set_ylabel('Distance From Home')

sns.barplot(data=df, x='WorkLifeBalance', y='Age', hue='Attrition', ax=axes3[2], palette=custom_palette, estimator=len)
axes3[2].set_title('Work-Life Balance Score (1=Low, 4=Best)')
axes3[2].set_ylabel('Employee Count')

plt.tight_layout()
plt.show()
```

```python
# Workplace Satisfaction & Culture Metrics
fig4, axes4 = plt.subplots(1, 3, figsize=(18, 5))
fig4.suptitle("Dashboard 4: Workplace Satisfaction & Culture Metrics", y=1.02, fontweight='bold', color='#1a1a1a')

sns.countplot(data=df, x='EnvironmentSatisfaction', hue='Attrition', ax=axes4[0], palette=custom_palette)
axes4[0].set_title('Environment Satisfaction (1-4)')

sns.countplot(data=df, x='JobSatisfaction', hue='Attrition', ax=axes4[1], palette=custom_palette)
axes4[1].set_title('Job Satisfaction (1-4)')

sns.countplot(data=df, x='RelationshipSatisfaction', hue='Attrition', ax=axes4[2], palette=custom_palette)
axes4[2].set_title('Relationship Satisfaction (1-4)')

plt.tight_layout()
plt.show()
```

```python

# Career Growth & Tenure Patterns
fig5, axes5 = plt.subplots(1, 3, figsize=(18, 5))
fig5.suptitle("Dashboard 5: Career Growth & Tenure Patterns", y=1.02, fontweight='bold', color='#1a1a1a')

sns.kdeplot(data=df, x='YearsAtCompany', hue='Attrition', fill=True, common_norm=False, ax=axes5[0], palette=custom_palette)
axes5[0].set_title('Tenure Distribution (Years at Company)')

sns.boxplot(data=df, x='Attrition', y='YearsSinceLastPromotion', ax=axes5[1], palette=custom_palette)
axes5[1].set_title('Years Since Last Promotion')

sns.violinplot(data=df, x='Attrition', y='YearsWithCurrManager', ax=axes5[2], palette=custom_palette)
axes5[2].set_title('Years with Current Manager')

plt.tight_layout()
plt.show()
```
